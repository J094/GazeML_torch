```python
import math
def compute_angular_loss(predict, label):
    """Pytorch method to calculate angular loss (via cosine similarity)"""
    def angle_to_unit_vectors(y):
        sin = torch.sin(y)
        cos = torch.cos(y)
        return torch.stack([
            cos[:, 0] * sin[:, 1],
            sin[:, 0],
            cos[:, 0] * cos[:, 1],
        ], dim=1)

    a = angle_to_unit_vectors(predict)
    b = angle_to_unit_vectors(label)
    ab = torch.sum(a*b, dim=1)
    a_norm = torch.sqrt(torch.sum(torch.square(a), dim=1))
    b_norm = torch.sqrt(torch.sum(torch.square(b), dim=1))
    cos_sim = ab / (a_norm * b_norm)
    cos_sim = torch.clip(cos_sim, -1.0 + 1e-6, 1.0 - 1e-6)
    ang = torch.acos(cos_sim) * 180. / math.pi
    return torch.mean(ang)
```

```python
dtype = torch.float
device = torch.device("cpu")
a = torch.randn(size=(32, 2), device=device, dtype=dtype, requires_grad=True)
b = torch.randn(size=(32, 2), device=device, dtype=dtype, requires_grad=True)
loss = compute_angular_loss(a, b)
```

```python
loss.backward()
```

```python
def compute_coord_loss(predict, label):
    loss = torch.nn.MSELoss()(predict, label)
    return loss
```

```python
dtype = torch.float
device = torch.device("cpu")
a = torch.randn(size=(32, 18, 2), device=device, dtype=dtype, requires_grad=True)
b = torch.randn(size=(32, 18, 2), device=device, dtype=dtype, requires_grad=True)
loss = compute_coord_loss(a, b)
```
